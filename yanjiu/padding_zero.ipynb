```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
a = np.arange(10)
signal_a = np.ones(10)
plt.plot(signal_a)
```

```python
A = np.fft.fft(signal_a)
plt.plot(np.abs(A))
```

```python
some_padding = np.zeros(15)
some_padding[:10] = signal_a
plt.plot(some_padding)
```

```python
A_some_padding = np.fft.fft(some_padding)
plt.plot(np.abs(A_some_padding))
```

```python
padding_1 = np.zeros(20)
padding_1[:10] = signal_a
plt.plot(padding_1)
```

```python
A_padding_1 = np.fft.fft(padding_1)
plt.plot(np.abs(A_padding_1))
```

```python
padding_zero = np.zeros(200)
padding_zero[:10] = signal_a
plt.plot(padding_zero)
```

```python
A_padding = np.fft.fft(padding_zero)
plt.plot(np.abs(A_padding))
```

```python
two_side = np.zeros(4000)
two_side[15:20] = 0.001
plt.plot(two_side)
```

```python
A_two_side = np.fft.fft(two_side)
plt.plot(np.abs(A_two_side))
two_side_padding = np.zeros(8000)
two_side_padding[0:4000] = np.abs(A_two_side)
two_side_padding[4000:8000] = np.abs(A_two_side)
plt.plot(np.abs(two_side_padding))
```
